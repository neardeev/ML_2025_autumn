```python
import io
from IPython.display import display
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

img = Image.open('1.jpg')
img = img.convert('RGB')

r, g, b = img.split()

r = np.array(r, dtype=np.float32)
g = np.array(g, dtype=np.float32)
b = np.array(b, dtype=np.float32)

U, S, V = np.linalg.svd(r, full_matrices= False)
U1, S1, V1 = np.linalg.svd(g, full_matrices= False)
U2, S2, V2 = np.linalg.svd(b, full_matrices= False)
r_pr = np.zeros_like(r, dtype = np.float32)
g_pr = np.zeros_like(g, dtype = np.float32)
b_pr = np.zeros_like(b, dtype = np.float32)





for i in range(1):
    r_pr += S[i] * np.outer(U[:,i], V[i, :])
    g_pr += S1[i] * np.outer(U1[:,i], V1[i, :])
    b_pr += S2[i] * np.outer(U2[:,i], V2[i, :])

r_pr[r_pr>255] =255
r_pr[r_pr<0] = 0

g_pr[g_pr>255] =255
g_pr[g_pr<0] = 0

b_pr[b_pr>255] =255
b_pr[b_pr<0] = 0

r_pr = r_pr.astype(np.uint8)
g_pr = g_pr.astype(np.uint8)
b_pr = b_pr.astype(np.uint8)

r_img = Image.fromarray(r_pr)
g_img = Image.fromarray(g_pr)
b_img = Image.fromarray(b_pr)

print(r_pr.dtype, g_pr.dtype, b_pr.dtype)

Y_r = Image.merge("RGB", (r_img, g_img, b_img))
display(Y_r)


buf = io.BytesIO()
Y_r.save(buf, format='JPEG')
print (len(buf.getvalue()))

img.save(buf, format='JPEG')
print (len(buf.getvalue()))
```
